# **Falta de acceso a internet en algunos departamentos de Colombia**

##*¿Que departamentos de Colombia actualmente tienen acceso a una red Wifi?*

Según los datos recolectados por el portal de Datos Abiertos Colombia, se pueden identificar los departamentos que ya cuentan con el beneficio de zonas Wi-Fi para la comunidad. Estos datos fueron obtenidos de diversas regiones de Colombia con el propósito de observar cuáles territorios tienen habilitados espacios donde la comunidad puede acceder a internet de manera gratuita y libre.






Con el siguiente codigo podemos observar toda la infomación recolectada de los departamentos, donde se evidencia la razon socia, la cual corresponde a los barrios donde se pueden encontrar zonas de acceso a Wi-Fi, ademas de su respectivos municipios y departamentos.

```python
df= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Proyecto/Zonas_Wi-Fi_20240911.csv')
print(df)
```

Vamos a filtar los datos para que sea mas sencillo manipular la infomación que necesitamos.

```python
columnas_seleccionadas = ['RAZON SOCIAL', 'NOMBRE_MUNICIPIO','NOMBRE_DEPARTAMENTO']
df_seleccionado = df[columnas_seleccionadas]
print(df_seleccionado)
```

Ahora con la infomación filtada creamos una grafica de barras para observar mejor los departamentos que cuentan con mayor cantidad de zonas Wi-Fi para la comunidad.

El gráfico permite visualizar el número de zonas Wi-Fi disponibles en cada departamento, destacando que Risaralda es uno de los territorios con más sitios de acceso a internet. Esto se desprende de los datos recolectados de los 26 departamentos de Colombia. En contraste, Arauca se presenta como la región con menos zonas Wi-Fi disponibles para la comunidad.

```python
datos = df['NOMBRE_DEPARTAMENTO'].value_counts()
ax = datos.plot(kind='bar', title="Frecuencia de Departamentos")
plt.xlabel("Nombre Departamentos")
plt.ylabel("Frecuencia")
plt.xticks(rotation=90)
plt.show()
```
